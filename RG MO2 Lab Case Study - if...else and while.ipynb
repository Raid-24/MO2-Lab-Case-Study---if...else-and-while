This Python app will accept student names and GPAs and test if the student made the Dean's List or Honor Roll.

```python
last_name = input("Enter student last name:")

while last_name != "ZZZ":
    first_name = input("Enter student first name:")
    student_gpa = float(input("Enter your GPA:"))

    if student_gpa >= 3.5:
        print("You made the Dean's List!")

    elif student_gpa >= 3.25:
        print("You made the Honor Roll!")
    
    last_name = input("Enter student last name:")
```
